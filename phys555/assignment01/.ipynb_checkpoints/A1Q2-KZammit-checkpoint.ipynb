## Introduction: SVD Theory

Singular value decomposition (SVD) is one of the most general and well used tools in linear algebra, and can be used for dimensionality reduction. 

For a large data set $x∈C^{nxm}$: 

X = \begin{bmatrix}
        | & | & \ & | \\
        x_{1} & x_{2} & \dots & x_{m} \\
        | & | & \ & |
    \end{bmatrix}
    
For face recognition, each column of X represents the pixels within an image. For example, if you have 400 faces, with 64x64 resolution, the X matrix would be 400 x $64*64$ dimensions. For the above matrix, there would be m faces in the dataset and n number of pixels in each image. 

SVD allows us to decompose X as a product of three other matrices: 
$X = UΣV^{T}$

where U (left singular vectors) and V (right singular vectors) are unitary (orthogonal) matrices, and Σ is diagonal (sigma are singular values, ordered from largest to smallest).

These are: 
U = 
    \begin{bmatrix}
        | & | & \ & | \\
        u_{1} & u_{2} & \dots & u_{m} \\
        | & | & \ & |
    \end{bmatrix}

Σ = 
\begin{bmatrix}
\sigma_{1} & 0 & \dots & 0 \\
0 & \sigma_{2} & \dots & 0 \\
\vdots & \vdots & \ddots & \vdots \\
0 & 0 & \dots & \sigma_{n} \\
0 & 0 & \vdots & 0 \\
\end{bmatrix}

$V^{T}$ = 
\begin{bmatrix}
    - & V_{1}^T & - \\
    - & V_{2}^T & - \\
    \ & \vdots &  \ \\
    - & V_{n}^T & - \\
\end{bmatrix}


The Σ matrix has at most m nonzero elements on the diagonal. The columns of U have the same shape as the columns of X. For the face recognition example, these can be considered as "eigenfaces". These eigenfaces are ordered hierarchically with respect to variance, meaning that the first eigenface describes the most variance in the columns of X. The relative importance of each eigenface is given by the corresponding singular value. This allows us to "chop off" unimportant columns from the end. 

Each V is a snapshot of how the corresponding U changes over the space. The first column of the transpose of V is an exact mixture of all columns of U to add them up to equal X1, scaled by the singular value.



## Introduction: SVD for Facial Recognition

Zeng, G. (2007) wrote a paper titled "Facial Recognition with Singular Value Decomposition" that in detail describes how SVD is used for facial recognition. As a summary, the steps to take are: 

1. Obtain a training set S with N faces 
2. Compute the mean face of set S, and form a matrix A of all faces minus the mean face 
3. Calculate SVD on matrix A 
4. Calculate the coordinate vector for each known individual 
5. Choose thresholds which define the maximum allowable distance from a known face in the training set 
6. For a new input image, calculate its coordinate vector, the vector projection, and the distance to the face space. 
7. If the distance to the face space is less than a threshold, calculate the distances to each face in the known space and take the minimum distance face to be the unknown individual

## SVD of Mariana

```python
# Using my friend Mariana as an example 

# Load her face in as a jpg
image = Image.open('mariana.jpg')

# Convert the image into an array 
image = img_to_array(image)

# Print the shape of the array
print("The shape of the original image is " + str(image.shape)) # (929, 697, 3)

# Calculate the mean of the array on axis 2 (this is because the image was read in as coloured even though it's black and white - could also do this) imgGray = color.rgb2gray(image) and take the mean of channel 1
img = np.mean(image,axis=2)
print("The shape of the grayscale image is " + str(img.shape))

#plt.imshow(img,'gray')
fig, (ax1, ax2) = plt.subplots(1,2)
plt.subplot(1,2,1)
plt.imshow(img,'gray')
plt.title('Original Mariana')

# Center the image at the mean
X_centered = img - img.mean()
print("The shape of the centered image is " + str(X_centered.shape))

# calculate U, sigma, and V for SVD
U, sigma, V = np.linalg.svd(X_centered)
print("The shape of U is " + str(U.shape)) # (929, 929)

# Using sigma components from a:b (ie. use the first 10 singular vectors and values)
a=0
b=10

# reconstruct the original matrix using U, V transpose, and sigma
reconstructed= np.matrix(U[:, a:b]) * np.diag(sigma[a:b]) * np.matrix(V[a:b, :])
plt.subplot(1,2,2)
plt.imshow(reconstructed,'gray')
plt.title('Reconstructed Mariana')
```

```python
# Define a function that plots the faces from the "images" variable
def plot_gallery(title, images, n_col, n_row, cmap=plt.cm.gray):
    
    # Set up the subplot formatting 
    fig, axs = plt.subplots(nrows=n_row, ncols=n_col, figsize=(2.0 * n_col, 2.3 * n_row), facecolor="white", constrained_layout=True)
    fig.set_constrained_layout_pads(w_pad=0.01, h_pad=0.02, hspace=0, wspace=0)
    fig.set_edgecolor("black")
    fig.suptitle(title, size=16)
    
    # For each image, plot it after reshaping it back into the original image shape
    for ax, vec in zip(axs.flat, images):
        vmax = max(vec.max(), -vec.min())
        im = ax.imshow(vec.reshape(image_shape), cmap=cmap, interpolation="nearest", vmin=-vmax, vmax=vmax)
        ax.axis("off")

    # Add a colourbar
    fig.colorbar(im, ax=axs, orientation="horizontal", shrink=0.99, aspect=40, pad=0.01)
    plt.show()
```

I will now plot 4 random faces from the Olivetti dataset, and then 4 eigenfaces created from PCA.

```python
# Initialize random number generator
rng = RandomState(0)

# Fetch the faces from the Olivetti dataset
faces, _ = fetch_olivetti_faces(return_X_y=True, shuffle=True, random_state=rng)

# Get the number of samples and features
n_samples, n_features = faces.shape

# Global centering (focus on one feature, centering all samples)
faces_centered = faces - faces.mean(axis=0)

# Local centering (focus on one sample, centering all features)
faces_centered -= faces_centered.mean(axis=1).reshape(n_samples, -1)

print("Dataset consists of %d faces" % n_samples)

# Set the number of faces to plot
n_row, n_col = 2, 2
n_components = n_row * n_col
image_shape = (64, 64)

# Plot the original faces 
plot_gallery("Faces from dataset", faces_centered[:n_components], n_col, n_row)

# Initialize PCA estimator
pca_estimator = decomposition.PCA(n_components=n_components, svd_solver="full", whiten=True)

# Apply PCA estimator on the centered faces 
pca_estimator.fit(faces_centered)

# Plot the resulting eigenfaces from the PCA
plot_gallery("Eigenfaces - PCA using randomized SVD", pca_estimator.components_[:n_components], n_col, n_row)
```

The eigenfaces are the corresponding columns of "U" to X. So, these are the "starting faces" which would be scaled by the corresponding singular value and mixture of "V", which is how these eigenfaces change over time to recreate the original X matrix.

## Bonus: Implementing Facial Recognition for SVD with the Olivetti Dataset

I closely followed the paper by Zeng, G. (2007), implementing the formulas and methods listed in the "Steps" section of their paper. 

The steps are as followed: 

1. Obtain a training set S with N face images of known individuals. An example of S is shown in Fig. 1. 
2. Compute the mean face f of set S by (2). The mean face of the set in Fig. 1 is shown in Fig. 2. Form matrix A in (4) with the computed f . 
3. Calculate the SVD of A as shown in (5). The result base-faces are shown in Fig. 3. 
4. For each known individual, compute the coordinate vector xi from (15). 
5. Choose a threshold 1 ε that defines the maximum allowable distance from face space. Determine a threshold 0 ε that defines the maximum allowable distance from any known face in the training set S. 
6. For a new input image f to be identified, calculate its coordinate vector x from (13), the vector projection p f , the distance f ε to the face space from (17). If f ε > 1 ε the input image is not a face. 
7. If f ε < 1 ε , compute the distance i ε to each known individual from (14). If all i ε > 0 ε , the input image may be classified as unknown face, and optionally used to begin a new individual face. If f ε < 1 ε , and some i ε < 0 ε , classify the input image as the known individual associated with the minimum i ε (xi) in (14), and this image may optionally added to the original training set. Steps 1-5 may be repeated. This can update the system with more instances of known faces. 

**The below code performs the SVD analysis on the Olivetti dataset.**

I selected a 25% test percentage, meaning that 75% of the data was used for training. Once SVD was performed, a randomly selected testing image is called and compared to the faces in the training dataset. Once the closest image is found (using the distance calculation), the training and test faces are plotted.

```python
from sklearn.model_selection import train_test_split

# Grab the faces from the Olivetti dataset 
faces = fetch_olivetti_faces()

# Split into test and train images 
img_train, img_test = train_test_split(faces.images, test_size=0.25, random_state=42)

#img = faces.images
print("The shape of the dataset is " + str(img_train.shape)) # (400, 64, 64), so 400 images that are 64x64

col_train = img_train.shape[1]*img_train.shape[2]
col_test = img_test.shape[1]*img_test.shape[2]

# Flatten into a 400 x (64x64) array
#reshaped = img.reshape(400,-1)
reshaped_train = img_train.reshape(img_train.shape[0], col_train) 
reshaped_test = img_test.reshape(img_test.shape[0], col_test)
print("The shape of the reshaped training dataset is " + str(reshaped_train.shape)) # (400, 4096)
print("The shape of the reshaped testing dataset is " + str(reshaped_test.shape))

# Compute the mean of the dataset 
mean_face = np.mean(reshaped_train, axis = 0)
mean_face = mean_face.reshape(1, reshaped_train.shape[1])
print('Mean face shape is ' + str(mean_face.shape))

# Subtract the mean from the dataset 
normalized = np.subtract(reshaped_train, mean_face)
normalized_test = np.subtract(reshaped_test, mean_face)
print('Normalized shape is ' + str(normalized.shape))

# Transpose this to ensure the following analysis are compatible
a_trans = normalized.transpose()
trans_test = normalized_test.transpose()
#print(a_trans.shape)

# Calculate SVD of A 
U, sigma, V = np.linalg.svd(a_trans)
print('U shape is ' + str(U.shape))

# Compute coordinate vector for each known individual 
xi = np.matmul(U.transpose(), a_trans)
print('coord vec shape is ' + str(xi.shape))

# Define thresholds 
epsilon_0 = 50
epsilon_1 = 10

## Classification
# Initialize distance array 
epsilons = np.zeros((img_train.shape[0],1))
print('epsilons shape is ' + str(epsilons.shape))

# test image and normalization 
# select a random image from the dataset 
n_pic = np.random.choice(np.arange(img_test.shape[0]))
test_image = trans_test[:,n_pic]
test_image = test_image.reshape(-1)
print('test image shape is ' + str(test_image.shape))

# Calculate coordinate vector of test image 
X = np.matmul(U.transpose(), test_image)
print('test image coord vec shape is ' + str(X.shape))

# Calculate final distance (formatted this way because I find it easier to read)
part1 = np.matmul(U, X)
epsilon_f = (np.matmul(np.subtract(test_image, part1).transpose(), (np.subtract(test_image, part1))))**0.5

# Check if the face is in the face space 
# if the distance is less than threshold one 
if epsilon_f < epsilon_1:
    # Compute the distance from each face 
    for ii in range(0, img_train.shape[0]):
        epsilons[ii,0] = np.matmul((xi[:,ii] - X).transpose(), (xi[:,ii] - X))
           
    # get the minimum distance and the index in epsilons at which that occurs 
    in_data = np.min(epsilons[:,0], axis=0)
    idex = np.argmin(epsilons[:,0], axis=0)
    #print(in_data)
    #print(idex)
    
    print("Above are shapes of matrices used in this analysis.")
    print("_________________________")
    
    if in_data < epsilon_0:
        print('The face belongs to person at index ' + str(idex) + ' in training dataset')

        
plt.subplot(1,2,1)
plt.imshow(trans_test[:,n_pic].reshape(64,64), 'gray')
plt.title('Training Dataset Image')

plt.subplot(1,2,2)
plt.imshow(a_trans[:,idex].reshape(64,64), 'gray')
plt.title('Identified Face')
        
```

## Conclusion

Singular value decomposition can be used to reduce the dimensions of a dataset. In terms of facial recognition, an image can be broken down into three representative matrices, which are intuitive to conceptualize. The U matrix represents the "eigenfaces", the Σ matrix represents the eigenvalues corresponding to the eigenfaces, and the columns of the V transpose represent the "mixture" needed to recreate the original data. 

The eigenfaces are called singular vectors and the corresponding values of sigma are called singular vectors. These are organized hierarchically in the matrix. The first vector and value represent the most important information from the original matrix, and the second the secondmost important, and so on. 

For Mariana, 10 vectors and values was not enough to recreate the original image, but 30 seemed to reconstruct the photo, mainting the amount of information that was in the original image. For the Olivetti dataset, 10 components was not enough to reconstruct the original image. Above 100 seems to be a good cutoff to represent the original data. 

For facial recognition, calculating the minimum "distance" of the coordinate vector of a test image to the coordinate vector of each known individual can determine which face within the training dataset is most similar to an individual. Changing the threshold values allows you to determine just *how* similar you'd like the images to be before considering them a match. Facial recognition was performed successfully for the Olivetti dataset, using a 75% train 25% test split.

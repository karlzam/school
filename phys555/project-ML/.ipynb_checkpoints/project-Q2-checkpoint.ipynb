# Functions

```python
# Confusion matrix function
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.viridis):
    
    #cmap=plt.cm.Blues
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    #print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.show()
```

# Load in Data and Balance Via Undersampling

```python
# Load in data and print out a few rows, as well as the shape of the data
df = pd.read_csv('covtype.csv')
print('The shape of the dataset is ' + str(df.shape))

# print the first few columns of the dataframe
df.head()
```

Looking at the above data, there is both binary and continuous data. For practice, I will use OneHot encoding for the cover type variable, even though it is already in numerical format.

```python
# Assign "cover type" as the target variable 
X = df.drop(["Cover_Type"], axis=1).values
y = df["Cover_Type"].values

# Print the number of samples within each class in the unaltered data
print('Original dataset shape %s' % Counter(y))  
```

As you can see, this is an imbalanced data set, where classes 1 and 2 have thousands more samples than the other classes. To avoid this issue (ie. ANN having bias towards predicting one class better than another), I have chosen to resample the data and use this sampled data going forward. There are many different ways to resample data, with some good options being KNN condensed neighbours to determine which samples to keep. For the sake of this assignment, I will use random undersampling as this dataset is quite large and my computing power is limited. 

```python
# Randomly undersample the data such that each class has the same samples as the minimum class
rus = RandomUnderSampler(random_state=42)
X_res, y_res = rus.fit_resample(X, y)
print('Resampled dataset shape %s' % Counter(y_res))
```

Now each class has balanced samples.

```python
# Practice using OneHotEncoder, although this is not necessary for this data set.
ohe = OneHotEncoder(sparse=False)
tar= np.reshape(y_res,(-1,1))
tar = ohe.fit_transform(tar)

# Split the data into training and validation, with 20% of the data for validation. 
X_tr, X_va, Y_tr, Y_va = train_test_split(X_res, tar, test_size=0.20, random_state=42)
print ('training set == ',np.shape(X_tr),np.shape(Y_tr),',, validation set == ', np.shape(X_va),np.shape(Y_va))

# Normalize the data using the StandardScaler function, ie. center the mean of the data around 0. 
scaler_S= StandardScaler().fit(X_tr)  # line #2
X_tr_Norm= scaler_S.transform(X_tr) # line # 3
X_va_Norm= scaler_S.transform(X_va)  # Line #4
```

# ANN

I now will use the Scikit Learn MLPClassifier to classify the data. Below is the pipeline I used to optimize the hyperparameters of the model, commented out for sake of time needed to rerun the notebook. Once I fine-tuned the optimized hyperparameters, I created a new classifier below with those predetermined values and fit the data to that. 

```python
# From previous pipelines: activation, solver, validation fraction, learning rate, hidden layer sizes 
# these were the best parameters before I tested alpha and tolerance

#ann_pipeline = Pipeline([('ANNcls', MLPClassifier(early_stopping=True, n_iter_no_change=5, max_iter=500,
#                                                  learning_rate='adaptive', solver='sgd', validation_fraction=0.1,
#                                                 hidden_layer_sizes=(20,50,20), activation='relu'))])
            
#params = [{'ANNcls__alpha':[1, 0.1, 0.01, 0.001], 'ANNcls__tol':[0.0001, 0.001, 0.01, 0.1]}]

#gs_ann = GridSearchCV(ann_pipeline,
#                      param_grid=params,
#                      scoring='accuracy',
#                      cv=10)

#gs_ann.fit(X_tr_Norm, Y_tr)

#gs_ann.best_params_
```

```python
# Create the ANN model with fine-tuned parameters 
clf_ANN = MLPClassifier(solver='adam', activation='relu', hidden_layer_sizes=(20,50,20), 
                        max_iter=500, early_stopping=True, n_iter_no_change=5, 
                        validation_fraction=0.1, learning_rate='adaptive', random_state=42, alpha=0.1, tol=0.0001)

## fitting the model:
clf_ANN.fit(X_tr_Norm, Y_tr)

## predict the response for tr and va sets. We can have two outputs: probability (e.g.,PY_tr_prob ) and  the winner class (e.g.,PY_tr):
ANN_tr_prob = clf_ANN.predict_proba(X_tr_Norm)
ANN_tr_pred = clf_ANN.predict(X_tr_Norm)

ANN_va_prob = clf_ANN.predict_proba(X_va_Norm)
ANN_va_pred = clf_ANN.predict(X_va_Norm)
```

```python
# Plot a confusion matrix for the training data 
cnf_matrix_ANN_tr = confusion_matrix(ANN_tr_pred.argmax(axis=1), Y_tr.argmax(axis=1))
plt.figure()
plot_confusion_matrix(cnf_matrix_ANN_tr, classes=['0','1', '2', '3', '4', '5', '6'], normalize=True,
                      title='Normalized Training Confusion Matrix')
```

From the training confusion matrix, we can conclude the network is good at predicting all classes (above 75% accuracy), but suffers when trying to predict class 1. Remembering that the initial data set had many more samples of class 1, perhaps this was intended as class 1 could be harder to predict, which would mean it would need more samples in order to accurately predict class 1. For the sake of this assignment, I will continue using the balanced data set. 

```python
# plot a confusion matrix for the validation data 
cnf_matrix_ANN_va = confusion_matrix(ANN_va_pred.argmax(axis=1), Y_va.argmax(axis=1))
plt.figure()
plot_confusion_matrix(cnf_matrix_ANN_va, classes=['0','1', '2', '3', '4', '5', '6'], normalize=True,
                      title='Normalized Validation Confusion Matrix')
```

The validation set does very similarly to the training data, which is a sign that the models hyperparameters were tuned accurately. Initially, one solution that came out of the pipeline lead to overfitting, which was detected because the validation set performed far worse than the training data. In this case, this seems to be a good model, and the best results I will see with the ANN. 

```python
# print the performance metrics of the model 
grid_predictions = clf_ANN.predict(X_va_Norm)
print(classification_report(Y_va.argmax(axis=1), grid_predictions.argmax(axis=1)))
```

Above are the performance metrics for each class of the data. Overall, the accuracy of the network is 71%, with a macro and weighted average of 72%. These are similar values that were achieved in the cited papers, and competitions that used this dataset. 

```python
# Compute ROC curve and ROC area for each class
fpr = dict()
tpr = dict()
roc_auc = dict()
for i in range(7):
    fpr[i], tpr[i], _ = roc_curve(Y_va[:, i], ANN_va_prob[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Compute micro-average ROC curve and ROC area
fpr["micro"], tpr["micro"], _ = roc_curve(Y_va.ravel(), ANN_va_prob.ravel())
roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])

# Plot ROC curve
plt.figure()
plt.plot(fpr["micro"], tpr["micro"],
         label='micro-average ROC curve (area = {0:0.2f})'
               ''.format(roc_auc["micro"]))
for i in range(7):
    plt.plot(fpr[i], tpr[i], label='ROC curve of class {0} (area = {1:0.2f})'
                                   ''.format(i, roc_auc[i]))

plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Multiclass Receiver Operating Characteristic Curves')
plt.legend(loc="lower right")
plt.show()
```

Above are the ROC curves for each class. It is shown here once more that the network classified class 1 the most incorrectly, but did a good job with the other classes.

```python
# Initialize the fine-tuned model
clf_svm = SVC(decision_function_shape='ovr', kernel='rbf', gamma=0.1, C=5, tol=1, probability=True, break_ties=True)

# Fit the model on the data 
clf_svm.fit(X_tr_Norm, Y_tr.argmax(axis=1))
```

```python
# Using those parameters, determine the training and validation predictions
SVM_tr_prob = clf_svm.predict_proba(X_tr_Norm)
SVM_tr_pred = clf_svm.predict(X_tr_Norm)

SVM_va_prob = clf_svm.predict_proba(X_va_Norm)
SVM_va_pred = clf_svm.predict(X_va_Norm)
```

```python
# Plot a confusion matrix for the training set 
cnf_matrix_tr_SVM = confusion_matrix(SVM_tr_pred, Y_tr.argmax(axis=1))
plt.figure()
plot_confusion_matrix(cnf_matrix_tr_SVM, classes=['0','1', '2', '3', '4', '5', '6'], normalize=True,
                      title='SVM - Normalized Confusion Matrix: Training')
```

Above is the confusion matrix for the training data from the SVC algorithm. Importantly, notice that class 1 has been more accurately predicted than possible with the fine-tuned ANN. Overall, these results have higher performance than the ANN. 

```python
# Plot a confusion matrix for the validation set
cnf_matrix_val_SVM = confusion_matrix(SVM_va_pred, Y_va.argmax(axis=1))
plt.figure()
plot_confusion_matrix(cnf_matrix_val_SVM, classes=['0','1', '2', '3', '4', '5', '6'], normalize=True,
                      title='SVM - Normalized Confusion Matrix: Validation')
```

The validation set performs very similarly to the training set, meaning that I do not expect that this model has suffered from overfitting. Overall, these results are better than the ANN. 

```python
# Plot the performance metrics of the SVC
grid_predictions = clf_svm.predict(X_va_Norm)
print(classification_report(Y_va.argmax(axis=1), grid_predictions))
```

Above are the performance metrics for each class of the data. Overall, the accuracy of the network is 82%, with a macro and weighted average of 82%.

```python
# Compute ROC curve and ROC area for each class
fpr = dict()
tpr = dict()
roc_auc = dict()
for i in range(7):
    fpr[i], tpr[i], _ = roc_curve(Y_va[:, i], SVM_va_prob[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Compute micro-average ROC curve and ROC area
fpr["micro"], tpr["micro"], _ = roc_curve(Y_va.ravel(), SVM_va_prob.ravel())
roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])

# Plot ROC curve
plt.figure()
plt.plot(fpr["micro"], tpr["micro"],
         label='micro-average ROC curve (area = {0:0.2f})'
               ''.format(roc_auc["micro"]))
for i in range(7):
    plt.plot(fpr[i], tpr[i], label='ROC curve of class {0} (area = {1:0.2f})'
                                   ''.format(i, roc_auc[i]))

plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Multiclass Receiver Operating Characteristic Curves')
plt.legend(loc="lower right")
plt.show()
```
